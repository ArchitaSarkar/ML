## MAE(Mean Absolute Error)

### Advantages:

- Intuitive Units: The resulting MAE value shares the same units as the actual output variable, making it straightforward to understand and explain.
- Outlier Robustness: MAE is less affected by extreme values or outliers in the data compared to other metrics like Mean Squared Error (MSE).
### Disadvantages:

- Non-Differentiable: The absolute value function used in MAE is not differentiable at zero. This characteristic can pose challenges for certain optimization algorithms, such as gradient descent, which rely on differentiability to efficiently find the best model parameters.

## MSE(Mean Square Error)

### Advantages:

- Differentiable: The squared function used in MSE is differentiable at all points. This makes it suitable for use as a loss function in many optimization techniques, such as gradient descent, which require a differentiable function to find optimal model parameters.
### Disadvantages:

- Unit Interpretation: The unit of MSE is the square of the output variable's unit, which can make it difficult to interpret in real-world terms. For example, if the output is in 'LPA', the MSE will be in 'LPA squared', which is not intuitively understandable.
- Sensitivity to Outliers: MSE heavily penalizes outliers because the errors are squared. Large errors from outliers become even larger when squared, causing them to disproportionately influence the overall error calculation and potentially skew the model's performance evaluation.

## RMSE

### Advantages:

- It is differentiable, making it suitable for use as a loss function in optimization algorithms like gradient descent.
The output unit is the same as the original - - output variable (e.g., if the output is in LPA, RMSE is also in LPA), which makes it easier to interpret compared to Mean Squared Error (MSE).
### Disadvantages:

- Similar to MSE, RMSE is sensitive to outliers, meaning that outliers can disproportionately impact the error calculation due to the squaring of errors.

## R2 Score

#### The R2 Score, also known as the Coefficient of Determination or Goodness of Fit, is a metric that indicates how well a regression model's predictions approximate the real data points. It essentially compares your regression line to a baseline model that simply predicts the mean of the target variable.

#### Here's a breakdown of the R2 Score:

### Formula and Calculation:

- The formula for R2 Score is 1 - (Sum of Squared Residuals of Regression Line / Sum of Squared Residuals of Mean Line).
- Sum of Squared Residuals of Regression Line (SSR): This measures the squared differences between the actual data points and the values predicted by your regression model.
- Sum of Squared Residuals of Mean Line (SSM): This measures the squared differences between the actual data points and the mean of the target variable.
### Interpretation of R2 Score:

- R2 Score = 0: This means your regression model performs no better than simply predicting the mean of the target variable. It suggests that your input features have no predictive power.
- R2 Score = 1: This indicates a perfect fit, where your regression line perfectly explains all the variance in the target variable, meaning there is no error in prediction. This is rarely achievable in real-world scenarios.
- R2 Score < 0 (Negative R2 Score): A negative R2 score signifies that your model performs worse than the mean line. This typically happens when the model is incorrectly applied to the data, for example, using a linear model on highly non-linear data.
- General Interpretation: An R2 score of, for example, 0.80 means that your input features explain 80% of the variance in the output column. The remaining percentage is unexplained variance, possibly due to other unconsidered factors or inherent randomness.
### Context Independence:

- Unlike error metrics like MAE, MSE, and RMSE, the R2 Score is context-independent. A good R2 score (e.g., 0.80) is generally considered good regardless of the specific dataset or units involved.

# Adjusted R Score

The Adjusted R Score is a modified version of the R Score that addresses a key flaw: the tendency of the regular R Score to increase when more input columns are added to a model, even if those columns are irrelevant. It penalizes the addition of features that do not significantly improve the model's explanatory power, making it a more reliable metric, especially in scenarios with multiple input features.

---

## Formula

Adjusted R = 1 - [(1 - R)  (n - 1) / (n - k - 1)]

Where:

- **R**  Normal R Score  
- **n**  Total number of rows (data points) in the dataset  
- **k**  Total number of independent input columns (features) in the model  

---

## How It Addresses the R Score's Flaw

The Adjusted R Score penalizes the addition of irrelevant features by considering the number of predictors (**k**) in the denominator of its formula.

### Adding an Irrelevant Column

- The normal R Score might slightly increase or stay the same.
- However, since **k increases**, the denominator (n - k - 1) decreases.
- This causes the overall Adjusted R Score to decrease.

This indicates that the added feature did not genuinely improve the model.

---

### Adding a Relevant Column

- The R Score increases significantly.
- Although **k increases**, the improvement in R is large enough to offset the penalty.
- As a result, the Adjusted R Score increases.

This confirms that the added feature is valuable to the model.

---

## Conclusion

The Adjusted R Score provides a more accurate assessment of model performance by considering model complexity (number of features) along with explanatory power.

If there is a significant difference between R and Adjusted R, it suggests that the R Score may be inflated due to irrelevant features.
