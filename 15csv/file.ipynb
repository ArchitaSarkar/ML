```python
# sep parameter
df3 = pd.read_csv('placement.csv', sep=';')
df3.head()
```

```python
pd.read_csv('placement.csv',names=['a','b','c','d','e']).head()   # when column name is not defined we give name explicitly
```

```python
pd.read_csv('placement.csv', index_col='cgpa')  # setting index column
```

```python
# header parameter
pd.read_csv('placement.csv', header=None).head()
```

```python
pd.read_csv('placement.csv', header=1).head() # skipping first row
```

```python
# use_cols parameter
pd.read_csv('placement.csv', usecols=['iq','placement']).head() # selecting specific columns
```

```python
# Squeeze parameter
iq = pd.read_csv('placement.csv', usecols=['iq']).squeeze() # to convert dataframe to series
iq.head()
```

```python
# skiprows/nrows parameter
print(pd.read_csv('placement.csv', skiprows=2).head())  # skipping first two rows
pd.read_csv('placement.csv', nrows=5).head()  # reading only first 5 rows
```

```python
# encoding parameter
pd.read_csv('placement.csv', encoding='utf-8').head()
```

```python
# dtype parameter
pd.read_csv('placement.csv', dtype={'iq': 'float64', 'cgpa': 'float64'}).dtypes
```

```python
# parse_dates parameter
# pd.read_csv('placement.csv', parse_dates=['status']).dtypes # converting status column to datetime format
```

```python
# converters parameter
pd.read_csv(
    'placement.csv',
    converters={'iq': lambda x: float(x) + 10}
).head()
```

```python
# na_values parameter
pd.read_csv('placement.csv', na_values={'placement': [0]}).head()
```

```python
# loading a huge file in chunks
dfs=pd.read_csv('placement.csv', chunksize=25)
for chunk in dfs:
    print(chunk.shape)
```
